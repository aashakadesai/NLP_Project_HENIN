```python
import json 

# generate unlabeled vine data from raw file
data = []
with open('sampled_post-comments_vine.json') as f:
    i = 0
    for line in f:
        if i > 20000:
            break
        data.append(json.loads(line.strip()))
        i += 1
```

```python
post_comment = {}
for comment in data:
    postId = comment['postId']
    if postId not in post_comment.keys():
        post_comment[postId] = []
    
    timestamp = comment['created'].split('T')
    out = '<font color="#0066CC">' + comment['username'] + '</font>   '\
        + comment['commentText'] + ' (created_at:' + timestamp[0] + ' '\
        + timestamp[1][:-7] + ')'
    post_comment[postId].append(out)
    
```

```python
lengths = []
for post in post_comment.keys():
    lengths.append(len(post_comment[post]))

print(max(lengths))
thresh = max(lengths)

for post in post_comment.keys():
    while len(post_comment[post]) < thresh:
        post_comment[post].append('empety')
```

```python
post_data = {}
with open('vine_posts_data.json') as f:
    for line in f:
        content = json.loads(line.strip())
        if content['_id'] in post_comment.keys():
            post_data[content['_id']] = content
```

```python
import pandas as pd

df_old = pd.read_csv('out.csv')
df_old.head()
```

```python
size = len(post_comment)
holder = [None for i in range(size)]
columns = {}
columns['_unit_id'] = [698432761 + i for i in range(size)]
columns['_golden'] = ['FALSE' for i in range(size)]
columns['_unit_state'] = ['finalized' for i in range(size)]
columns['_trusted_judgments'] = ['5' for i in range(size)]
columns['_last_judgment_at'] = ['4/4/2015 2:22:09 AM' for i in range(size)]
columns['question1'] = list(df_old['question1'])
columns['question1:confidence'] = [1 for i in range(size)]
columns['question2'] = list(df_old['question2'])
columns['question2:confidence'] = [1 for i in range(size)]

for i in range(thresh):
    columns['clmn'+ str(i)] = holder.copy()

columns['cptn_time'] = holder.copy()
columns['owner_cmnt'] = holder.copy()
columns['id'] = holder.copy()
columns['owner_id'] = holder.copy()
columns['img_url'] = holder.copy()
columns['shared media'] = [0 for i in range(size)]
columns['followed_by'] = [0 for i in range(size)]
columns['follows'] = [0 for i in range(size)]
columns['likes'] = holder.copy()
```

```python
out_df = pd.DataFrame(columns)
out_df.head()
```

```python
# add all comments:
clmn = [x for x in out_df.columns if 'clmn' in x] # column of comments
i = 0
for post in post_comment.keys():
    cmnts = post_comment[post]
    for j in range(len(clmn)):
        cl = clmn[j]
        out_df.loc[i, cl] = cmnts[j]
    
    if post not in post_data.keys():
        post = '1001397703332573184'
    p_dat = post_data[post]
    timestamp = p_dat['created'].split('T')
    out_df.loc[i, 'id'] = post
    out_df.loc[i, 'cptn_time'] = 'Media posted at ' + timestamp[0] + ' ' + timestamp[1][:-7]
    out_df.loc[i, 'owner_cmnt'] = p_dat['description']
    out_df.loc[i, 'owner_id'] ='<font color="#0066CC">' + p_dat['username'] + '</font>'
    out_df.loc[i, 'img_url'] = p_dat['shareUrl']
    out_df.loc[i, 'likes'] = p_dat['likeCount'] + '   likes'
    i += 1
```

```python
out_df.to_csv('out2.csv', index=False)
```
